Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.


Answer: Descriptive statistics summarize data you already have. For example, calculating the average age of survey participants.
Inferential statistics help you make predictions about a larger group using a sample. Like estimating national preferences from a small survey.


Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.


Answer: Sampling means picking a subset from a population to study.
Random sampling gives everyone an equal chance. Stratified sampling splits the population into groups and samples from each—ensuring fair representation.


Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.


Answer:
•	Mean is the average.
•	Median is the middle value.
•	Mode is the most frequent value.
•	They help describe the center of your data and spot patterns.


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?


Answer: Skewness shows if data leans left or right. A positive skew means most values are low, with a few high outliers.
Kurtosis tells how sharp the peak is. High kurtosis means more extreme values.
